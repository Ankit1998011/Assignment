```python
# OverflowError:-

#Ans. An OverflowError exception is raised when an arithmetic operation exceeds the limits to be represented. This is part of the ArithmeticError Exception class. 

import math

logging.info("Solution to Q3")
logging.error("Q3. OverFlowError Example")
print(math.exp(1000))
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

Ans. Lookup Error acts as a base class for the exceptions that occur when a key or index used on a mapping or sequence of a list/dictionary is invalid or does not exists. 
The two types of exceptions raised are: IndexError. KeyError.

```python
""" KeyError:- A Python KeyError exception is what is raised when you try to access a key that isn't in a dictionary ( dict ). 
Python's official documentation says that the KeyError is raised when a mapping key is accessed and isn't found in the mapping.
A mapping is a data structure that maps one set of values to another."""

# Example:-
logging.info("Solution to Q.4")
logging.error("Q4. KeyError Example")
d1 = {'a':1,'b':2}

print(d1['c'])
```

```python
# IndexError:- This error occurs when an attempt is made to access an item in a list or tuple at an index which is out of bounds. 

# Example:-
logging.error("Q4. IndexError Example")

l1 = [1,2,3,4,5,6]

print(l1[10])
```

Q5. Explain ImportError. What is ModuleNotFoundError?

Ans. ImportError:- It is raised when an import statement has trouble successfully importing the specified module.
Typically, such a problem is due to an invalid or incorrect path, which will raise a ModuleNotFoundError in Python 3.6 and newer versions.

ModuleNotFoundError:- this error occurs when you're trying to access or use a module that cannot be found.

Q6. List down some best practices for exception handling in python.

Ans. Some best practices for exception handling are:-

1.Use try/catch/finally blocks to recover from errors or release resources.

2.Handle common conditions without throwing exceptions.

3.Throw exceptions instead of returning an error code.

4.Use always a specific exception.

5.Print always a proper message.

6.Always try to log your error

7.Document all the error.

8.Cleanup all the resources
