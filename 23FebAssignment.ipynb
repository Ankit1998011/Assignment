Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

Ans. Pandas DataFrame is a two-dimensional structure with columns and rows whose size can be changed.
We can also think of it as a dictionary of Series objects. 

while,
Pandas Series is a one-dimensional labeled array whose object size cannot be changed.
we can also see it as the primary building block for a DataFrame, making up its rows and columns. 

```python
#example:-

# DataFrame:-

logging.info("Q4.DataFrame Example")
df = pd.DataFrame({"Name":['Alice','Bob','Claire'],
                   "Age":[25,30,27],
                   "Gender":['Female','Male','Female']
})

print(type(df))
df
```

```python
# Series is basically building block of DataFrame:-

logging.info("Q4.Series Example")
print(type(df['Name']))
df['Name']
```

Q6. Which of the following is mutable in nature Series, DataFrame, Panel?

Ans. All pandas data structures are value-mutable (the values they contain can be altered) but not always size-mutable. The size of a Series cannot be changed, but, for example, columns can be inserted into a DataFrame.

In Serie  data mutable but size immutable.
In both DataFrame and Panel  data and size are mutable.
